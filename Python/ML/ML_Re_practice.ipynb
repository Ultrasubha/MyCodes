```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_percapita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year,df.per_capita_income_ussd,color='green', marker='^')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income_ussd)
```

```python
reg.predict(2020)
```
