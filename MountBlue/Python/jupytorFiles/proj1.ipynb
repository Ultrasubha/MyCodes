<h1>Q.1) Total runs scored by team</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt

dT = pd.read_csv("deliveries.csv")
dT
```

```python
totalRuns = dT.groupby('batting_team')['total_runs'].sum().reset_index().sort_values(by = 'total_runs', ascending = False)
totalRuns
```

```python
plt.figure(figsize=(12,6))
plt.bar(totalRuns['batting_team'],totalRuns['total_runs'],color='red')
plt.xticks(rotation=45,ha='right')
plt.title('Total Runs Scored by Each IPL Team')
plt.xlabel('Teams')
plt.ylabel('total_runs')
plt.tight_layout()
plt.show()
```

```python
matches = pd.read_csv("matches.csv")
matches.head()
```

<h1>Q.4) Stacked chart of matches played by team by season</h1>

```python
games_by_team = matches.groupby('team1').size() + matches.groupby('team2').size()
games_by_season = matches.groupby('season').size()

stacked_data = pd.DataFrame({'Team': games_by_team.index, 'Games Played': games_by_team.values})
stacked_data_pivot = matches.pivot_table(index='season', columns='team1', aggfunc='size', fill_value=0)
stacked_data_pivot = stacked_data_pivot.add(matches.pivot_table(index='season', columns='team2', aggfunc='size', fill_value=0), fill_value=0)

stacked_data.plot(kind='bar', x='Team', y='Games Played', stacked=True, figsize=(12, 6))
plt.title('Number of Games Played by Team')
plt.xlabel('Team')
plt.ylabel('Number of Games Played')
plt.xticks(rotation=45,ha='right')
plt.show()

stacked_data_pivot.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Number of Games Played by Season')
plt.xlabel('Season')
plt.ylabel('Number of Games Played')
plt.legend(title='Team', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45,ha='right')
plt.show()
```

<h1>Q.5) Number of matches played per year for all the years in IPL</h1>

```python
matches_per_year = matches['season'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
matches_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Played per Year in IPL')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

<h1>Q.6) Number of matches won per team per year in IPL</h1>

```python
matches_won_per_team_per_year = matches.groupby(['season', 'winner']).size().unstack()

plt.figure(figsize=(12, 8))
matches_won_per_team_per_year.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Number of Matches Won per Team per Year in IPL')
plt.xlabel('Year')
plt.ylabel('Number of Matches Won')
plt.legend(title='Team', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<h1>Q.7) Extra runs conceded per team in the year 2016</h1>

```python
data_2016 = matches[matches['season'] == 2016].copy() #copying so that modifications don't affect original DataFrame
data_2016['extra_runs_conceded'] = 0
#for matches won
data_2016.loc[data_2016['win_by_runs'] > 0, 'extra_runs_conceded'] = data_2016['win_by_runs']
#for matches won
data_2016.loc[data_2016['win_by_wickets'] > 0, 'extra_runs_conceded'] = data_2016['win_by_wickets'] * 6
extra_runs_per_team = data_2016.groupby('team1')['extra_runs_conceded'].sum()
extra_runs_per_team
```

```python
plt.figure(figsize=(10, 6))
extra_runs_per_team.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Extra Runs Conceded per Team in IPL 2016')
plt.xlabel('Team')
plt.ylabel('Extra Runs Conceded')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<h1> Q.8) Top 10 economical bowlers in the year 2015 </h1>

```python
merged_data = pd.merge(matches, dT, left_on='id', right_on='match_id', how='inner')
data_2015 = merged_data[merged_data['season'] == 2015].copy()

# Economy rate per bowler
data_2015['total_runs'] = data_2015['total_runs'].astype(int)
economy_data = data_2015.groupby('bowler').agg({'total_runs': 'sum', 'ball': 'count'})
economy_data['economy'] = (economy_data['total_runs'] / economy_data['ball']) * 6

top_10_economical_bowlers = economy_data.sort_values(by='economy').head(10)

plt.figure(figsize=(10, 6))
top_10_economical_bowlers['economy'].plot(kind='bar', color='skyblue')
plt.title('Top 10 Economical Bowlers in IPL 2015')
plt.xlabel('Bowler')
plt.ylabel('Economy Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
